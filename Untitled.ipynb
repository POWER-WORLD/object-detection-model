```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
config_file = 'C:/Users/manishdanu07/python/first project/ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'C:/Users/manishdanu07/python/first project/frozen_inference_graph.pb'
```

```python
model = cv2.dnn_DetectionModel(frozen_model,config_file)
```

```python
classlabels = []
file_name = 'C:/Users/manishdanu07/python/first project/Labels.txt'
with open(file_name,'rt') as fpt:
    classLabels = fpt.read().rstrip('\n').split('\n')
    
```

```python
print(classLabels)
```

```python
print(len(classLabels))
```

```python
model.setInputSize(320, 320)  # Set input size for the model
model.setInputScale(1.0 / 127.5)  # Set input scale
model.setInputMean((127.5, 127.5, 127.5))  # Set input mean for normalization
model.setInputSwapRB(True) 
```

```python
cap = cv2.VideoCapture(1)
# Check if the video is opened correctLy
if not cap.isOpened() :
    cap = cv2.VideoCapture(0)
if not cap.isOpened() :
    raise IOError("Cannot open webcam")

font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret,frame = cap.read()
    ClassIndex, confidece, bbox = model.detect (frame, confThreshold=0.55)

    print(ClassIndex)
    if (len(ClassIndex)!=0):
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
            if (ClassInd<=80):
                cv2.rectangle(frame, boxes, (255, 0, 0), 2 )
                cv2.putText (frame, classLabels [ClassInd-1], (boxes [0]+10, boxes [1]+40), font, fontScale=font_scale, color= (0, 255, 0),thickness=3)

    cv2.imshow('0bject Detection Tutorial',frame)

    if cv2.waitKey(2) & 0XFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()            
```
